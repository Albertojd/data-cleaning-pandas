```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(('CleanShark.csv'), engine = 'python')
df.head()
```

```python
#### Intentaremos ver si en los años en los que la temperatura del mar ha sido más elevada han aumentado significativamente los ataques
#### Nos centramos en Florida y New South Wales, los lugares con más ataques
```

```python
df["Area"].value_counts().nlargest(5).plot(kind="bar")
```

```python
df_flo = df[(df.Year >= 1928)&(df.Year <= 2016)&(df.Area == 'Florida')]
```

```python
df_flo["Year"].value_counts().nlargest(20).plot(kind="bar")
plt.savefig("FloridaAttacks.PNG", bbox_inches='tight')
```

![title](FloridaWater.png)

```python
df_nsw = df[(df.Year >= 1928)&(df.Year <= 2010)&(df.Area == 'New South Wales')]
```

```python
df_nsw["Year"].value_counts().nlargest(20).plot(kind="bar")
```

![title](Tasmansea.png)

```python
%matplotlib inline
df.Area.value_counts().nlargest(5).plot.pie(labels=['FLORIDA', 'NEW SOUTH WALES', 'QUEENSLAND', 'HAWAII', 'CALIFORNIA'], 
 fontsize=15, figsize=(6, 6), colors = ('#9dbd00','#66b3ff','#db900d','#ffcc99','#009dbd'),
                                                    explode=(0.15, 0, 0, 0, 0), title = 'Areas with more shark attacks')
plt.savefig("Prueba.jpg", bbox_inches='tight')
```
